classifier = Sequential()

classifier.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu'))

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_train = pd.read_csv("titanic/train.csv")
```

```python
data_train.drop(['PassengerId','Name'],axis=1,inplace=True)
```

```python
data_train.Sex.replace({'male':0,'female':1},inplace=True)
```

```python
data_train.Age.fillna(data_train.Age.mean(),inplace=True)
```

```python
data_train.Embarked.replace({np.NaN:'S'},inplace=True)
```

```python
data_train.drop('Cabin',axis=1,inplace=True)
```

```python
data_train.drop('Ticket',axis=1,inplace=True)
```

```python
X = data_train.iloc[:,1:].values
y = data_train.iloc[:, 0].values
```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder

labelencoder_embarked = LabelEncoder()
X[:, -1] = labelencoder_embarked.fit_transform(X[:, -1])
```

```python
onehotencoder = OneHotEncoder(categorical_features = [6])
X = onehotencoder.fit_transform(X).toarray()
```

```python
X = X[:,1:]
```

```python

# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X = sc.fit_transform(X)

```

```python
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 42)
```

```python

# Importing the Keras libraries and packages
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
X_train[1]
```

```python

# Initialising the ANN
classifier = Sequential()

# Adding the input layer and the first hidden layer
classifier.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu', input_dim = 8))

# Adding the second hidden layer
classifier.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu'))

# Adding the output layer
classifier.add(Dense(units = 1, kernel_initializer = 'uniform', activation = 'sigmoid'))

# Compiling the ANN
classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python

# Fitting the ANN to the Training set
classifier.fit(X_train, y_train, batch_size = 10, epochs = 100)
```

```python
y_pred = classifier.predict(X_test)
```

```python
y_pred = (y_pred > 0.5)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
cm = confusion_matrix(y_test, y_pred)
acc = accuracy_score(y_test,y_pred)
```
